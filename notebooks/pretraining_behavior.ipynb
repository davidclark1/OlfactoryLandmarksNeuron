## Generated figures:
* S1F (pre-training behavior: licking and speed)


```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import sem
import warnings

from os import sys
sys.path.append('..')
import data_util
import plotting_util
import style

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['svg.fonttype'] = "none"

FIG_DIR = "../figures/"
```

```python
lick_rate = data_util.load_all_binned_behavior_data("lick_rate")
vel = data_util.load_all_binned_behavior_data("vel")
```

```python
#Use factor of 1e3 to convert lick rate from 1/ms to 1/s
lick_rate_mouse_means = np.nanmean(lick_rate, axis=(2, 3)) * 1e3
vel_mouse_means = np.nanmean(vel, axis=(2, 3))
```

```python
x = (np.arange(40) + .5)*.1
y = np.array([np.swapaxes(lick_rate_mouse_means, 0, 1),
              np.swapaxes(vel_mouse_means, 0, 1)])
titles = ["Day " + str(d) for d in np.arange(-2, 6)]
xlabel = None
ylabels = ["Lick Rate\n(1/s)", "Speed\n(cm/s)"]


plotting_util.plot_mean_and_ind_lines_grid(x, y, titles, xlabel, ylabels,
                                           figsize=(6, 1.2), hspace=.5, xlim=[0, 4], xticks=[0, 4])

plt.savefig(FIG_DIR + "pretraining_behavior.svg", bbox_inches="tight")
```
